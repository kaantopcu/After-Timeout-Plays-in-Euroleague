```python
# Drop unnecessary columsn
df = df.loc[:, df.columns[3:]]
df
```

```python
playTypes = df[['PLAYTYPE','PLAYINFO']].drop_duplicates(subset='PLAYTYPE')
playTypes
```

```python
Game_Info = df[['Game ID','Team_A','Team_B']].copy()
Game_Info
```

```python
Game_Info = Game_Info.dropna().copy()
Game_Info = Game_Info.drop_duplicates().copy()
Game_Info
```

```python
# Eventlist of free throws made, field goals made and asssist to be used to detect assist and related score event
#eventList = ['FTA','FTM','2FGA','2FGM','3FGA','3FGM','TOUT','TOUT_TV','TO','EP']
eventList = ['2FGA','2FGM','3FGA','3FGM','TOUT','TO']

# Filtering only plays in the eventlist
timeoutAfterPlays = df[df['PLAYTYPE'].isin(eventList)]
timeoutAfterPlays.drop_duplicates(inplace = True)
```

```python
# Shots are kept in the previous plays of the assists. Finding previous plays (shots made of the assists)
timeoutAfterPlays['followingNUMBEROFPLAY'] = timeoutAfterPlays['NUMBEROFPLAY'].shift(-1)
timeoutAfterPlays
```

```python
# Keeping only the asssists and their correspondence previous events and renaming them
timeouts = timeoutAfterPlays[(timeoutAfterPlays['PLAYTYPE'] == "TOUT")]
#timeouts.rename(columns = {'prevNUMBEROFPLAY': 'shotEventNumber', 'NUMBEROFPLAY': 'assistEventNumber'}, inplace = True)
timeouts
```

```python
# Keep relevant columns
timeoutsWithPlays = timeouts[['Game ID','NUMBEROFPLAY','followingNUMBEROFPLAY','CODETEAM']]
timeoutsWithPlays
```

```python
# Merging the shots df with the whole ply by play data to get the relevant information
afterTimeoutPlays = pd.merge(timeoutsWithPlays,df, how ='left', left_on = ['Game ID','followingNUMBEROFPLAY','CODETEAM'], right_on = ['Game ID','NUMBEROFPLAY','CODETEAM'])
afterTimeoutPlays
```

```python
# There are some cases that two consecutive assists are played in the data, to eliminate that plays that have assists in their previous plays are filtered out
afterTimeoutPlays = afterTimeoutPlays[afterTimeoutPlays['PLAYTYPE'] != "TOUT"]
#afterTimeoutPlays = afterTimeoutPlays[afterTimeoutPlays['PLAYTYPE'] != "TOUT_TV"]
afterTimeoutPlays
```

```python
# Count the number of each events for each team
teamEvents = afterTimeoutPlays.groupby('TeamName').PLAYTYPE.value_counts()
#team_events = merged_data.groupby('TEAM_NAME').EVENTMSGTYPE_PLAY.value_counts().reset_index(name='playCount')
#team_events = merged_data.groupby('TEAM_NAME').EVENTMSGTYPE_PLAY.value_counts()
teamEvents
```

```python
# Calculating related event counts for each team

twoPointMadeCount =  teamEvents[:, "2FGM"]
threePointMadeCount = teamEvents[:, "3FGM"]

twoPointMissCount =  teamEvents[:, "2FGA"]
threePointMissCount = teamEvents[:, "3FGA"]

missedShotCount = twoPointMissCount + threePointMissCount
madeShotCount = twoPointMadeCount + threePointMadeCount

turnoversCount = teamEvents[:,"TO"]

allShotCount = madeShotCount + missedShotCount

allActionCount = allShotCount + turnoversCount

teamEff = madeShotCount / allShotCount

teamAllEff = madeShotCount / allActionCount

madePct = madeShotCount / allActionCount

missedPct = missedShotCount / allActionCount

toPct = turnoversCount / allActionCount
```

```python
# Winning Pct of the last season
winPct = teamStats[["Team","TeamName", "WPCT"]]
winPct
```

```python
# Converting team efficiency to dataframe and append winning percentage
teamEfficiency = teamAllEff.to_frame()
teamEfficiencyWithWinPct = pd.merge(left=winPct, right=teamEfficiency, left_on="TeamName",right_on='TeamName')

# Multiply winning pct and efficiency by 100
teamEfficiencyWithWinPct.WPCT = teamEfficiencyWithWinPct.WPCT * 100
teamEfficiencyWithWinPct.PLAYTYPE  = teamEfficiencyWithWinPct.PLAYTYPE  * 100

teamEfficiencyWithWinPct = teamEfficiencyWithWinPct[['TeamName','WPCT','PLAYTYPE']]
teamEfficiencyWithWinPct
```

```python
madeShots = afterTimeoutPlays.loc[(afterTimeoutPlays.PLAYTYPE == "2FGM") | (afterTimeoutPlays.PLAYTYPE == "3FGM")]
missedShots = afterTimeoutPlays.loc[(afterTimeoutPlays.PLAYTYPE == "2FGA") | (afterTimeoutPlays.PLAYTYPE == "3FGA")]
madeShots
```

```python
# Count the number of shots by each player
shotMadeCountsId = madeShots.PLAYER.value_counts()
shotMissedCountsId = missedShots.PLAYER.value_counts()
shotMissedCountsId
```

```python
shotsMadeDf = shotMadeCountsId.to_frame(name="MADE_COUNT")
shotsMissedDf = shotMissedCountsId.to_frame(name="MISSED_COUNT")
shotsMissedDf
```

```python
#shots_made_frame.reset_index(level=0, inplace=True)
shotsMadeDf['PLAYER'] = shotsMadeDf.index
shotsMadeDf = shotsMadeDf.reset_index(level=0, drop=True)

shotsMissedDf['PLAYER'] = shotsMissedDf.index
shotsMissedDf = shotsMissedDf.reset_index(level=0, drop=True)



mergedShots = pd.merge(shotsMadeDf, shotsMissedDf)
mergedShots
```

```python
afterTimeOutExecutors = pd.merge(left=mergedShots, right=afterTimeoutPlays.loc[:, ['PLAYER']], on='PLAYER', how='left').drop_duplicates()
afterTimeOutExecutors
```

```python
afterTimeOutExecutors.loc[:, "TOTAL_COUNT"] = afterTimeOutExecutors.MADE_COUNT + afterTimeOutExecutors.MISSED_COUNT
afterTimeOutExecutors
```

```python
afterTimeOutExecutors = afterTimeOutExecutors.sort_values(by='TOTAL_COUNT', ascending=False)
```

```python
afterTimeOutExecutors.loc[:, "PERCENTAGE"] = afterTimeOutExecutors.MADE_COUNT / afterTimeOutExecutors.TOTAL_COUNT
```

```python
topExecutors = afterTimeOutExecutors.iloc[:100]
```

```python
topTwenty = topExecutors.iloc[:20]
```
